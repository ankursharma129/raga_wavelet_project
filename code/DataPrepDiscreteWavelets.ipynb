```python
import pandas as pd
import os
import librosa
import scipy
import numpy as np
wavelets = ['db4', 'coif10', 'rbio2.4', 'bior1.5']
# wavelets = ['sym16']
filePath = "./saraga1.5_hindustani"
csv_path = os.path.join(filePath, "file_paths.csv")
df = pd.read_csv(csv_path)
raga_filenames = []

for i in range(108):
    raga_filenames.append(os.path.join(filePath, df["filepath"][i])+".mp3.mp3")
    
pPath = "./saraga1.5_hindustani/"
paths = []
for i in df["filepath"]:
    mod_path = i.split("/")
    print(pPath+mod_path[0]+"/"+mod_path[1])
    paths.append(pPath+mod_path[0]+"/"+mod_path[1])
for wave in wavelets:
    files = []
    for fold in paths:
        tag = fold.split('/')[-1]
        for file in os.listdir(fold):
            if file.endswith(wave+"level7.pickle"):
#                 print(file)
                full_file_path = fold+"/"+file
                files.append(full_file_path)
    dat = []
    for i in files:
        if i.split("/")[-2] not in dat:
            dat.append(i.split("/")[-2])
    lab = {}

    for i in range(84):
        tag = dat[i]
        y_label = np.zeros((84))
        y_label[i] = 1
        lab[tag] = y_label
    data = [[file, file.split("/")[-2]] for file in files]
    header = ["file", "tag"]
    import csv
    with open('dataGen'+wave+'level7.csv', 'w', encoding='UTF8', newline='') as f:
        writer = csv.writer(f)

        # write the header
        writer.writerow(header)

        # write multiple rows
        writer.writerows(data)
```
